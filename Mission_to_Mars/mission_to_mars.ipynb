```python
from bs4 import BeautifulSoup as bs
import requests
from splinter import Browser
import pandas as pd
```

```python
#URL to be scraped
#url = 'https://redplanetscience.com/' This didn't work after trying numerous times
url = 'https://mars.nasa.gov/news/' #This was recommended on Slack
response = requests.get(url)
```

```python
# Create BeautifulSoup object; parse with 'html.parser'
soup = bs(response.text, 'html.parser')
type(soup)
```

```python
print(soup.prettify())
```

```python
#Narrow the div and class to see how best to loop through
results = soup.body.find_all('div', class_="content_title")
results
```

```python
# A blank list to hold the headlines
news_titles = []
# Loop over div elements
for result in results:
    # Identify the anchor...
    if (result.a):
        # And the anchor has non-blank text...
        if (result.a.text):
            # Append thext to the list
            news_titles.append(result)
news_titles
```

```python
finalnewstitles = []
# Print only the headlines
for x in range(6):
    var=news_titles[x].text
    newvar = var.strip('\n\n')
    finalnewstitles.append(newvar)
finalnewstitles
```

```python
#Find classification for description paragraph below title
presults = soup.find_all('div', class_="rollover_description_inner")
presults
```

```python
news_p = []
# Loop through the div results to pull out just the text
for x in range(6):
    var=presults[x].text
    newvar = var.strip('\n\n')
    news_p.append(newvar)
news_p
```
